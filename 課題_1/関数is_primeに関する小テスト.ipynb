## 小テスト　1a

素数を判定する関数is_primeを以下に作りました。
（なお，素数とは，1より大きい自然数で，約数が1と自分自身のみである数です）
そして，この関数を使って，5は素数であり，10は素数でないことを確認しています。

しかし，このis_prime関数には誤りがあるので，正しく動作するように修正してください。
ただし，別のプログラムに置き換えてはいけません。エラー箇所を修正すること。

```python
#　小テスト 1a
# 素数か否かを判定する関数
def is_prime(number):
    """入力値が素数ならTrueを、
       素数でなければFalseを返す"""
    # 1以下は素数ではない
    if number <= 1:   
        return False
    # 2から自分自身未満の数で割り切れば素数ではない
    for i in range(2, number):
        if number % i == 0:     
            return False        
    return True

# 5は素数であり，10は素数でないことを確認する
print(is_prime(5))    # True
print(is_prime(10))   # False
```

## 小テスト　1c

このis_prime関数を使って，2から50までの素数を要素とする配列（list）を作成し，表示しなさい

```python
# 小テスト 1c ここにプログラムを記述

def is_prime(number):
    
    if number <= 1:   
        return False

    for i in range(2, number):
        if number % i == 0:     
            return False        
    return True

prime_numbers = [num for num in range(2, 51) if is_prime(num)]

print("2から50までの素数リスト:")
print(prime_numbers)
```
